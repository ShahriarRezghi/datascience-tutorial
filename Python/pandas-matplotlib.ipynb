```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
A = np.random.rand(20)
B = np.random.rand(20)
A, B
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(A, B)
ax.set_title("My figure of random numbers")
ax.set_xlabel("My X Label")
ax.set_ylabel("My Y Label")
fig.tight_layout()
fig.savefig('my-figure.pdf', dpi=400)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.scatter(A, B, label='Normal Dots')
ax.scatter(A+1, B+1, label='Shifted Dots')
ax.legend()
fig.tight_layout()
```

```python
A = np.random.normal(size=1000)
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.hist(A, bins=100)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.bar(['A', 'B', 'C'], [10, 20, 15])
fig.tight_layout()
```

```python
A = np.random.normal(size=1000)
B = np.random.normal(size=1000)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].hist(A)
ax[1].bar(['A', 'B', 'C'], [10, 20, 15])
fig.tight_layout()
```

```python
A = np.random.normal(size=(100, 3))
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.boxplot(A)
fig.tight_layout()
```
