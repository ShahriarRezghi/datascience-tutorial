```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv("diamonds.csv")
table
```

```python
del table['Unnamed: 0']
```

```python
table['volume'] = table['x'] * table['y'] * table['z']
table
```

```python
table.describe()
```

```python
plt.hist(table.volume)
plt.show();
```

```python
table[table['volume']>500].sort_values(by=['volume','z'],ascending=False)
```

```python
plt.hist(table[table['volume']<1000].volume)
plt.show();
```

```python
plt.scatter(table.x, table.y)
plt.xlabel('x')
plt.ylabel('y')
plt.show();
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(table['x'], table['y'], table['z'])
plt.show();
```

```python
A = table[table['cut'] == 'Premium']
B = table[table['cut'] == 'Good']
A
```

```python
plt.scatter(A.x, A.y)
plt.scatter(B.x, B.y)
plt.show();
```
